```python
import pandas as pd

def find_large_text_columns(df, min_length=1000):
    """
    Find columns in the DataFrame where the text length in the first few rows is greater than min_length.

    :param df: DataFrame to check.
    :param min_length: Minimum length of text to be considered as 'large'.
    :return: List of column names that contain large text.
    """
    large_text_columns = []
    for column in df.columns:
        # Check the length of the text in the first few rows of the column
        sample_texts = df[column].astype(str).head(10)  # Sample first 10 rows
        if any(len(text) > min_length for text in sample_texts):
            large_text_columns.append(column)
    return large_text_columns

# Path to your CSV files
csv_paths = [
    '/content/drive/My Drive/Sydhur/CSV1.csv',
    '/content/drive/My Drive/Sydhur/CSV2.csv',
    '/content/drive/My Drive/Sydhur/CSV3.csv',
    '/content/drive/My Drive/Sydhur/CSV4.csv'
]

# Check each CSV file
for csv_path in csv_paths:
    df = pd.read_csv(csv_path)
    large_text_columns = find_large_text_columns(df)
    print(f'Large text columns in {csv_path}: {large_text_columns}')
```

# **Task 1**

```python
import pandas as pd

# Define the paths to your CSV files and the names of the large text columns
csv_files = {
    '/content/drive/My Drive/Sydhur/CSV1.csv': 'SHORT-TEXT',
    '/content/drive/My Drive/Sydhur/CSV2.csv': 'entites',
    '/content/drive/My Drive/Sydhur/CSV3.csv': 'TEXT',
    '/content/drive/My Drive/Sydhur/CSV4.csv': 'TEXT'
}

# Path to the output .txt file
output_txt_path = '/content/drive/My Drive/Sydhur/combined_text.txt'

# Initialize an empty list to collect text
all_texts = []

# Process each CSV file
for csv_path, text_column in csv_files.items():
    df = pd.read_csv(csv_path)
    # Extract text from the specified column
    texts = df[text_column].astype(str).tolist()
    all_texts.extend(texts)

# Combine all text into a single string with newlines
combined_text = "\n".join(all_texts)

# Write the combined text to a .txt file
with open(output_txt_path, 'w') as file:
    file.write(combined_text)

print(f"Text from all CSV files has been combined and saved to {output_txt_path}")
```

**3.1 Count Word Occurrences**

```python
import pandas as pd
from collections import Counter
import re

# Load the text data from the .txt file
with open('/content/drive/My Drive/Sydhur/combined_text.txt', 'r') as file:
    text = file.read()

# Tokenize the text into words (simple whitespace-based tokenization)
words = re.findall(r'\b\w+\b', text.lower())  # Convert to lower case and find words

# Count word occurrences
word_counts = Counter(words)

# Get the top 30 most common words
top_30_words = word_counts.most_common(30)

# Convert to DataFrame and save as CSV
df_top_30_words = pd.DataFrame(top_30_words, columns=['Word', 'Count'])
df_top_30_words.to_csv('/content/drive/My Drive/Sydhur/top_30_words.csv', index=False)

print("Top 30 words and their counts have been saved to 'top_30_words.csv'.")
```

```python
import pandas as pd
from collections import Counter
import re

# Load the text data from the .txt file
with open('/content/drive/My Drive/Sydhur/combined_text.txt', 'r') as file:
    text = file.read()

# Tokenize the text into words, excluding numbers
words = re.findall(r'\b[a-zA-Z]+\b', text.lower())  # Only matches alphabetic words

# Count word occurrences
word_counts = Counter(words)

# Get the top 30 most common words
top_30_words = word_counts.most_common(30)

# Convert to DataFrame and save as CSV
df_top_30_words = pd.DataFrame(top_30_words, columns=['Word', 'Count'])
df_top_30_words.to_csv('/content/drive/My Drive/Sydhur/top_30_words_1.csv', index=False)

print("Top 30 words and their counts (excluding numbers) have been saved to 'top_30_words.csv'.")
```

**3.2: Count Unique Tokens Using Transformers**

```python
from transformers import AutoTokenizer
from collections import Counter
import pandas as pd

# Load the text data from the .txt file
with open('/content/drive/My Drive/Sydhur/combined_text.txt', 'r') as file:
    text = file.read()

# Load the BioBERT tokenizer
tokenizer = AutoTokenizer.from_pretrained('dmis-lab/biobert-v1.1')

# Tokenize the text
tokens = tokenizer.tokenize(text)

# Count token occurrences
token_counts = Counter(tokens)

# Get the top 30 most common tokens
top_30_tokens = token_counts.most_common(30)

# Convert to DataFrame and save as CSV
df_top_30_tokens = pd.DataFrame(top_30_tokens, columns=['Token', 'Count'])
df_top_30_tokens.to_csv('/content/drive/My Drive/Sydhur/top_30_tokens.csv', index=False)

print("Top 30 tokens and their counts have been saved to 'top_30_tokens.csv'.")
```

```python
from transformers import BertTokenizer, BertForTokenClassification, pipeline
import torch

# Load BioBERT model and tokenizer
biobert_model_name = "dmis-lab/biobert-base-cased-v1.1"
tokenizer = BertTokenizer.from_pretrained(biobert_model_name)
biobert_model = BertForTokenClassification.from_pretrained(biobert_model_name)
nlp_ner = pipeline("ner", model=biobert_model, tokenizer=tokenizer)

# Example text (replace with actual text loading)
text = "Sample text with diseases and drugs."

# Extract entities using BioBERT
def extract_entities_biobert(text):
    ner_results = nlp_ner(text)
    entities = {result['word'] for result in ner_results}
    return entities

entities_biobert = extract_entities_biobert(text)
print("Entities extracted by BioBERT:", entities_biobert)
```

```python
import spacy

# Load the model
nlp_sci = spacy.load('en_core_sci_sm')
nlp_bc5cdr = spacy.load('en_ner_bc5cdr_md')

# Process text
doc_sci = nlp_sci(text)
doc_bc5cdr = nlp_bc5cdr(text)

# Extract entities
entities_sci = [ent.text for ent in doc_sci.ents]
entities_bc5cdr = [ent.text for ent in doc_bc5cdr.ents]

# Count entities
from collections import Counter
entities_sci_counts = Counter(entities_sci)
entities_bc5cdr_counts = Counter(entities_bc5cdr)

# Compare results
common_entities = set(entities_sci_counts) & set(entities_bc5cdr_counts)
diff_entities_sci = set(entities_sci_counts) - set(entities_bc5cdr_counts)
diff_entities_bc5cdr = set(entities_bc5cdr_counts) - set(entities_sci_counts)

# Save results
with open('entities_comparison.txt', 'w') as f:
    f.write("Entities detected by SciSpaCy:\n")
    f.write(str(entities_sci_counts) + '\n')
    f.write("Entities detected by BC5CDR:\n")
    f.write(str(entities_bc5cdr_counts) + '\n')
    f.write("Common Entities:\n")
    f.write(str(common_entities) + '\n')
    f.write("Entities in SciSpaCy but not in BC5CDR:\n")
    f.write(str(diff_entities_sci) + '\n')
    f.write("Entities in BC5CDR but not in SciSpaCy:\n")
    f.write(str(diff_entities_bc5cdr) + '\n')
```
